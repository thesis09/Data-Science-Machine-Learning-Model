```python
x=d.iloc[:,2:13].values
y=d.iloc[:,-2].values
```

```python
x
```

```python
y
```

**Creating a linear progession object**

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict([[67.450578,118.127103,0.97,15.69,16.46,23.483476,0.97,24.55,34.06,3.68000,5.500000]])
```

**AQI Predection is done here**

```python
a=model.coef_
b=model.intercept_
y_pred=a*x+b
y_pred
```
